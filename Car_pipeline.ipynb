```python
def frame_summ(dataframe)->DataFrame:
    summary=[]
    for i in dataframe.columns:
        summary.append(len(dataframe[i].unique()))
    data=pd.DataFrame({'unique_val':summary},index=dataframe.columns)
    data['d_types']=dataframe.dtypes
    return data
    
```
